#Q1.Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
df['profit'] = df['revenue'] - df['budget']
df.head()
```

```python
max_pro_movie = df[df['profit'] == df['profit'].max()]['title'].values[0]
max_pro_movie
```

```python
max_movie = df.loc[df['profit'].idxmax()]
max_movie
```

```python
eval(max_movie['cast'])
```

```python
for i in eval(max_movie['cast']):
  print(i['name'])
```

```python
actor_list = []
for i in eval(max_movie['cast']):
  actor_list.append(i['name'])
print(actor_list)
```

```python
max_movie['crew']
```

```python
Director_list = []
for i in eval(max_movie['crew']):
  if i['job'] == 'Director':
    Director_list.append(i['name'])

Producer_list = []
for i in eval(max_movie['crew']):
  if i['job'] == 'Producer':
    Producer_list.append(i['name'])
print(Producer_list)
print(Director_list)
```

```python
print('the max profitable movie is :- ', max_pro_movie)
print('actor in that movie :-', actor_list)
print('producer in that movie :-', Producer_list)
print('director in that movie :-', Director_list)
```

#Q3.Find out the unique genres of movies in this dataset.

```python
df['genres'][1:5]
```

```python
# This will throw an error due ti null values

for i in df['genres'][0:]:
  for j in eval(i):
    print(j['name'])
```

```python
df.info()
```

```python
not_null_genres = df[df['genres'].notnull()]
not_null_genres.isnull().sum()
```

```python
genere_list = []
for i in not_null_genres['genres'][0:]:
  for j in eval(i):
    genere_list.append(j['name'])
print(genere_list)
```

```python
set(genere_list)
```

```python
genere_df = pd.DataFrame(genere_list,columns = ['genres'])
genere_df.head()
```
