```python
###### OOPS ######
class bpit:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def student_name(self):
    print(f"this is class bpit and student name is: {self.name}")
  def student_age(self):
    print(f"this is class bpit and student age is: {self.age}")

obj = bpit('raghav',10)
obj.student_name()
obj.student_age()


class radius:
  def __init__(self,r):
    self.r = r
  def circle(self):
    print(3.14*self.r*self.r)
obj = radius(10)
obj.circle()
```

```python
# anagram --> in this remove extra element to make the array same {order is not important}

a = list('cde')
b = list('dcf')
u_a = []
u_b = []
delete = []
print(a)
print(b)

for i in a:
  for j in b:
    if(a[0:len(i)-1]==b[0:len(j)-1]):
      b.pop()
      u_a.append(i)
      u_b.append(j)
      # a.pop()
    else:
      delete.remove(i)
      delete.remove(j)
print(set(u_a))
print(set(u_b))
print(a)
print(b)
print(delete)
```
