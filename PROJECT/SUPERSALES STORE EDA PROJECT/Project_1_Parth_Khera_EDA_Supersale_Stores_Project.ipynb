# **Project Summary -**

The exploratory data analysis (EDA) of the Superstore dataset provides meaningful insights into sales performance, customer behavior, product demand, and regional contribution. The dataset consists of multiple dimensions such as customer details, product information, order dates, shipping details, and sales values, which together help in understanding overall business performance.

To begin with, significant data preprocessing and feature engineering were performed. The **Order_Date and Ship_Date columns were converted from object type to datetime format**, enabling proper time-series analysis. From these date columns, several new features were created, including **Order_Year, Order_Month, Order_Day, Ship_Year, Ship_Month, Ship_Day, and Shipping_Days**. These derived features helped in analyzing seasonal trends, delivery efficiency, and yearly performance patterns more effectively.

The analysis of sales distribution across regions shows that **sales are not evenly distributed**, with certain regions contributing a higher share of revenue compared to others. This indicates a strong regional imbalance in business performance. Similarly, segment-wise analysis reveals that the **Consumer segment dominates in terms of order volume and sales contribution**, while Corporate and Home Office segments contribute less comparatively. This suggests that the business is heavily dependent on a single customer segment.

Customer-level analysis highlights that a **small number of top customers contribute a large portion of total sales**, indicating a high dependency on key customers. This presents both an opportunity and a risk. While retaining these customers can ensure stable revenue, losing them could significantly impact overall sales performance.

Product-level analysis shows that **sales are concentrated in a few top-performing products and sub-categories**, while many others contribute relatively less. This indicates uneven product demand distribution and suggests the need for better product strategy, inventory management, and targeted promotions for underperforming items.

From a temporal perspective, the analysis shows that **sales fluctuate across months and years**, indicating seasonal trends in customer purchasing behavior. Certain months show higher sales activity, which can be used for planning promotions and stock management. However, the **correlation analysis reveals that numerical variables such as Order Year, Ship Year, and Shipping Days have very weak relationships with Sales**, suggesting that time-based factors alone do not strongly influence revenue.

The shipping analysis further indicates that **Shipping Days do not significantly affect sales**, meaning customers are not highly sensitive to delivery duration within the observed range. This implies that logistics performance is not a major driver of sales variation in this dataset.

Overall, the EDA reveals that the business is primarily driven by **a combination of region, customer segment, and product category**, rather than purely numerical or time-based factors. There is a clear imbalance where a small set of regions, customers, and products contribute disproportionately to revenue. While this concentration helps identify key growth drivers, it also introduces risk due to dependency on limited sources of income.

In conclusion, the Superstore dataset analysis highlights the importance of **targeted business strategies such as focusing on high-performing regions, retaining valuable customers, optimizing top products, and improving underperforming segments**. Leveraging these insights can help improve profitability, reduce risk, and ensure sustainable long-term business growth.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df.dtypes
````

````python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format = "%d/%m/%Y")
````

````python
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format = "%d/%m/%Y")
````

````python
df.dtypes
````

````python
# Write your code to make your dataset analysis ready.
df.info()
````

````python
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month_name()
df['Order_Day'] = df['Order_Date'].dt.day_name
````

````python
df['Ship_Year'] = df['Ship_Date'].dt.year
df['Ship_Month'] = df['Ship_Date'].dt.month_name()
df['Ship_Day'] = df['Ship_Date'].dt.day_name
````

````python
df['Shipping_Days'] = (df['Ship_Date'] - df['Order_Date']).dt.days
````

````python
df.info()
````

### What all manipulations have you done and insights you found?

The manipulations done are as follow:
1. Order_Date and Ship_Date columns data types have been changed form object to datetime64.
2. Seven new columns have been created from Order_Date and Ship_Date:
*   Order_Year
*   Order_Month
*   Order_Day
*   Ship_Year
*   Ship_Month
*   Ship_Day
*   Shipping_Days

The insights are as follow:

1. Sales are unevenly distributed across regions, segments, and states, with some areas contributing significantly more revenue.
2. A small group of top customers and products contribute a major share of total sales, showing dependency on key drivers.
3. Shipping days have little impact on sales, indicating customers are not highly sensitive to delivery time in this dataset.
4. Sales show seasonal variation across months and years, helping identify peak and low-performing periods.
5. Numerical variables alone show weak correlation with sales, meaning categorical factors are more important for business decisions.
